```r
indCon <- function(vectorCliente) {
    serieMedias = rep(1,96)
    serieMedias[1:24]= mean(vectorCliente[1:24])
    serieMedias[25:48]= mean(vectorCliente[24:48])
    serieMedias[49:72]= mean(vectorCliente[48:72])
    serieMedias[73:96]= mean(vectorCliente[72:96])
    indc=vectorCliente/serieMedias
    
 }
```

```r
indCon1 <- function(vectorCliente) {
    serieMedias = rep(1,72)
    serieMedias[1:24]= mean(vectorCliente[1:24])
    serieMedias[25:48]= mean(vectorCliente[24:48])
    serieMedias[49:72]= mean(vectorCliente[48:72])
    indc=vectorCliente/serieMedias
    
 }
```

```r
indRes = function(vectorTorre,poblacion){
indr <- vector()
for(num in 1:96){ 
        slicingTorre = vectorTorre[1:num] 
        metric = (log(sum(slicingTorre)/num))/log(poblacion)
        indr[num]=metric  
          }
return(as.numeric(indr))
}
```

```r
indRes1 = function(vectorTorre,poblacion){
indr <- vector()
for(num in 1:72){ 
        slicingTorre = vectorTorre[1:num] 
        metric = (log(sum(slicingTorre)/num))/log(poblacion)
        indr[num]=metric  
          }
return(as.numeric(indr))
}
```
