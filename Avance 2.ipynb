```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('EventosManabi.csv')
df['FechaEvento'] = pd.to_datetime(df['FechaEvento'])
```

```python
df
```

```python
def categorizarHora(x):
    if x.Hora >= 0 and x.Hora<=6:
        return 'MADRUGADA'
    elif x.Hora > 6 and x.Hora<=12:
        return 'MANANA'
    elif x.Hora > 12 and x.Hora<=18:
        return 'TARDE'
    elif x.Hora > 18 and x.Hora<=23:
        return 'NOCHE'
```

```python
df['Tiempo'] = df.apply(categorizarHora, axis=1)
```

```python
df
```

```python
df.groupby('Tiempo').size().sort_values(ascending=False).plot('bar', title='Numero total de eventos por canton')
```

```python
dfantes = df.loc[df['FechaEvento']<'2016-04-19']
dfdespues = df.loc[df['FechaEvento']>'2016-04-19']
```

```python
numtorres = dfantes.groupby(['CantonEvento', 'FechaEvento', 'Tiempo'])['BaseEvento'].nunique()
totaleventos = dfantes.groupby(['CantonEvento', 'FechaEvento', 'Tiempo']).size()
dfantesagrupado = pd.concat([numtorres, totaleventos], axis=1).rename(columns={"BaseEvento": "NumeroTorresActivas", 0: "TotaEventos"})
dfantesagrupado = dfantesagrupado.reset_index()
dfantesagrupado
```

```python
dfantesagrupado.corr()
```

```python
g = sns.FacetGrid(dfantesagrupado, col='Tiempo',row="FechaEvento", hue='CantonEvento', aspect=1.75)
g = g.map(plt.scatter, "NumeroTorresActivas","TotaEventos",edgecolor="w").add_legend()
```
